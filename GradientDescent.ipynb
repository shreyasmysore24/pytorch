```python
import numpy as np
import sklearn
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=4, n_features=1,n_informative=1,noise=13, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

```python
m=model.coef_
model.intercept_,m,X,y
```

```python
slope=-2*np.sum(y-m*X.ravel()-0)
b=slope
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),c='r')
```

```python
for i in range(100):
    b=b-(0.1)*(-2*np.sum(y-m*X.ravel()-b))
b
```

```python
y_pred=m*X+b

plt.scatter(X, y, label='Data', color='black')  # Actual data points
plt.plot(X, model.predict(X), c='r', label='Model Prediction')  # Model's prediction
plt.plot(X, y_pred, c='b', label='Our Calculation')  # Our manual prediction

# # Labels and legend
# plt.xlabel('X')
# plt.ylabel('y')
plt.legend()
plt.title('Linear Regression Prediction')
plt.show()
```
